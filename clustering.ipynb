# Elbow Test

```python
# Create a list to store inertia values and the values of k
inertia = []

# Create a list to set the range of k values to test
k = list(range(1, 11))
```

```python
# Create a for loop where each value of k is evaluated using the K-means algorithm
# Fit the model using the "customers_df" DataFrame
# Append the value of the computed inertia from the `inertia_` attribute of the KMeans model instance
for i in k:
    k_model = KMeans(n_clusters=i, random_state=0)
    k_model.fit(df_medical_scaled)
    inertia.append(k_model.inertia_)
```

```python
# Define a DataFrame to hold the values for k and the corresponding inertia
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

# Review the DataFrame
df_elbow.head()
```

```python
# Plot the DataFrame to identify the optimal value for k
df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

### Step 4: Initialize the K-means model with three clusters and then fit the `df_medical_scaled` DataFrame to the model.

```python
# Initialize the K-Means model with n_clusters=3
model = KMeans(n_clusters=2)
```

```python
# Fit the model for the df_stocks_scaled DataFrame
model.fit(df_medical_scaled)
```

### Step 5. Predict the clusters and then create a new DataFrame with the predicted clusters.

```python
# Predict the model segments (clusters)
medical_clusters = model.predict(df_medical_scaled)

# View the stock segments
print(medical_clusters)
```

```python
# Create a new column in the DataFrame with the predicted clusters
df_medical_scaled["MedicalCluster"] = medical_clusters

# Review the DataFrame
df_medical_scaled.head()
```

### Step 6: Reduce the number of features to two principal components on the `df_medical_scaled` DataFrame, and  calculate the explained variance ratio that results from the PCA data.

```python
# Create the PCA model instance where n_components=2
pca = PCA(n_components=3)
```

```python
# Fit the df_stocks_scaled data to the PCA
medical_pca_data = pca.fit_transform(df_medical_scaled)

# Review the first five rose of the PCA data
# using bracket notation ([0:5])
medical_pca_data[:5]
```

```python
# Calculate the explained variance
pca.explained_variance_ratio_
```

### Step 7: Use the calculate PCA DataFrame in Step 6 to create a new DataFrame called, `df_medical_pca`, then add an additional column to the `df_medical_pca` DataFrame that contains the tickers from the original `medical_df` DataFrame.

```python
# Creating a DataFrame with the PCA data
df_medical_pca = pd.DataFrame(medical_pca_data, columns=["PC1", "PC2", "PC3"])

# Copy the tickers names from the original data
df_medical_pca["Zipcode"] = medical_df.index

# Set the Ticker column as index
df_medical_pca = df_medical_pca.set_index("Zipcode")

# Review the DataFrame
df_medical_pca.head()
```

### Step 8: Rerun the K-means algorithm on the `df_medical_pca` DataFrame and create a scatter plot using the  "MedicalCluster" and the two principal components for the x- and y-axes. Be sure to style and format your plot.

```python
# Initialize the K-Means model with n_clusters=3
model = KMeans(n_clusters=2)

# Fit the model for the df_stocks_pca DataFrame
model.fit(df_medical_pca)

# Predict the model segments (clusters)
medical_clusters = model.predict(df_medical_pca)

# Print the stock segments
print(medical_clusters)
```

```python
# Create a copy of the df_stocks_pca DataFrame and name it as df_stocks_pca_predictions
df_medical_pca_predictions = df_medical_pca.copy()

# Create a new column in the DataFrame with the predicted clusters
df_medical_pca_predictions["MedicalCluster"] = medical_clusters

# Review the DataFrame
df_medical_pca_predictions.head()
```

```python
# Create the scatter plot with x="PC1" and y="PC2"
df_medical_pca_predictions.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="MedicalCluster",
    title = "Scatter Plot by Stock Segment - PCA=2"
)
```
